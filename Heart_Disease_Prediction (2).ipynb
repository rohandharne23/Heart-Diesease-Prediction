```python
#DATA VISUALIZATION

#Creating age groups
data['age'].unique()

#Ages range from 32 to 70 so the following age groups will be created:
   # 30-39 years
   # 40-49 years
   # 50-59 years
   # 60-70 years

data['ageGroup'] = pd.cut(x=data['age'], bins=[30, 39, 49, 59, 70], labels=['30-39 years', '40-49 years', '50-59 years', '60-70 years'])
data.head()
```

```python
#Checking for correlation among individual independent variables
correlations = data.corr()
correlations

#plotting correlation on HEAT MAP
import seaborn as sns
f, ax = plt.subplots(figsize=(10,8))
sns.heatmap(correlations, annot=True,cmap="YlGnBu");
```

```python
#The correlation coefficients of the variables individually did not shown a strong correlation with TenYearCHD.
#This only confirms that the propensity of developing heart disease is a result of a conjunction 
#of factors.
```

```python
#STACKED BAR GRAPH
Fig=pd.crosstab(data.ageGroup,data.TenYearCHD)
Fig.div(Fig.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Age-Wise Chances of Heart Disease')
plt.xlabel('Age')
plt.ylabel('Heart Disease')
plt.legend(["No Heart Diseases", "Have heart diseases"])

```

```python
#Ages above 50 have more chances of having heart diseases
```

```python
df=data[['male','age','currentSmoker','BPMeds','prevalentStroke','cigsPerDay','diabetes','totChol','heartRate']]
df.head()
```

```python
#Now converting the variables into dependent and independent variables
```

```python
#Defining Independent Variables
X = df.copy()
X
```

```python
#Defining Dependent Variable
y = data['TenYearCHD']
y
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# spliting data as 70% for training and 30% as testing data
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.30, random_state = 99)

train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
#Import LogisticRegression
from sklearn.linear_model import LogisticRegression

#Instantiate a LogisticRegression classifier
lr = LogisticRegression(solver = 'liblinear')
```

```python
#Fitting lr to train set
lr.fit(train_x, train_y)
```

```python
#Use lr to predict instances from the test set and store it
y_pred = lr.predict(test_x)
y_pred
```

```python
#Checking Accuracy of model
lr.score(test_x, test_y)

# Get the accuracy score of logreg model and print it
print("Accuracy of logistic regression classifier: ", lr.score(test_x, test_y))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix
accuracy_score(test_y, y_pred)
```

```python
#Building CONFUSION MATRIX
confusion_matrix(test_y, y_pred)
```

```python
#PLotting a CONFUSION MATRIX
plot_confusion_matrix(lr, test_x, test_y,cmap = "YlGnBu")
```
